```python
class WeatherAPI:
    
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = 'https://api.weatherapi.com/v1'
        
    def _get(self, q='', api_method='/current.json'):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={q}"
        res = requests.get(request_url)
        if res.ok:
            return res.json()
        else:
            print('There was an error')
    
    def get_current_weather(self, city):
        data = self._get(city)
        display_name = data['location']['name'] + ', ' + data['location']['region']
        current_temp = data['current']['temp_f']
        current_condition = data['current']['condition']['text']
        feels_like = data['current']['feelslike_f']
        degree_sign = u"\N{DEGREE SIGN}F"
        print(f"It is currently {current_condition} and {current_temp}{degree_sign} in {display_name}. It feels like {feels_like}{degree_sign}.")


weather = WeatherAPI('f72035084b9043abb7d52225212212')         
```

#### In-class Exercise

Add a method for getting the forecast weather for a city. It should default to 3 days in the future. It should display the high, low and condition with the date.

```python
class WeatherAPI():
    def __init__(self, api_key):
        self.api_key = api_key
        self.base_url = 'http://api.weatherapi.com/v1'
    
    # single '_' means the function will be used internally ONLY
    def _get(self, q='', api_method='/current.json', days=''):
        request_url = self.base_url + api_method + '?key=' + self.api_key + '&q=' + q
        
        # Attaches days to the end of the url only if "days" has a value
        if days != '':
            request_url += f'&days={days}'
            
        r = requests.get(request_url)
        if r.ok:
            return r.json()
        else:
            print('There was an error')
    
    def get_current_weather(self, city):
        # Have to use self.<function> for it to work in a class
        data = self._get(city)
        city_location = data['location']['name'] + ', ' + data['location']['region']
        current_temp = data['current']['temp_f']
        current_condition = data['current']['condition']['text']
        feels_like = data['current']['feelslike_f']
        degree_sign = u"\N{DEGREE SIGN}F"
    
    def get_current_weather(self, city):
        # Have to use self.<function> for it to work in a class
        data = self._get(city)
        city_location = data['location']['name'] + ', ' + data['location']['region']
        current_temp = data['current']['temp_f']
        current_condition = data['current']['condition']['text']
        feels_like = data['current']['feelslike_f']
        degree_sign = u"\N{DEGREE SIGN}F"
        
        print(f"It is currently {current_condition} and {current_temp}{degree_sign} in {city_location} and it feels like {feels_like}{degree_sign}")   
    
    def get_forecast_weather(self, city, days=3):
        data = self._get(city, '/forecast.json', days)
        degree_sign = u"\N{DEGREE SIGN}F"
        
        city_location = data['location']['name'] + ', ' + data['location']['region']
        print(city_location)
        print('='*50)
        
        # Get the dates and the high, low and condition for each day
        for day in data['forecast']['forecastday']:
            print('Date:', day['date'])
            print('High:', day['day']['maxtemp_f'], degree_sign)
            print('Low:', day['day']['mintemp_f'], degree_sign)
            print('Condition:', day['day']['condition']['text'])
            print('='*50)
```

```python
def run():
    weather_checker = WeatherAPI('9215f22b1dc34f078c8193322212212')
    while True:
        ask = input('For which city would you like to see the weather or "quit" to exit? : ').lower()
        
        if ask == 'quit':
            break
            
        option = input('Would you like to see the "current" or "forecast" weather?: ').lower()
        
        if option == 'current':
            weather_checker.get_current_weather(ask)
        elif option == 'forecast':
            weather_checker.get_forecast_weather(ask)
        else:
            print('Please enter "current" or "forecast."')
run()              


```

#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
class Poke_HW():
    def __init__(self):
        self.base_url = 'https://pokeapi.co/api/v2/pokemon/'
       
    def _get(self, name):
        request_url = self.base_url + name
        res = requests.get(request_url)
    
        if res.ok:
            return res.json()
        else:
            print('There was an error.')
        
    def get_info(self, name):
        data = self._get(name)
        
        print(name.title(), 'is', data['height'],'feet tall and', data['weight'],'lbs')
        print('-' * 50)
```

```python
def run():
    while True:
        ask = input('Which pokemon would you like to see or "quit" to exit? :').lower()
        
        if ask == 'quit':
            break
        
        Poke_HW().get_info(ask)
        
run()
```
